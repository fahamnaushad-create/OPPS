##1. What is Object-Oriented Programming (OOP)?

  - OOP is a programming style where we organize code using objects that contain data and functions together. It helps in breaking large programs into smaller reusable parts. OOP makes programs easier to maintain and scale. Python fully supports OOP through classes and objects.

##2. What is a class in OOP?

- A class is a blueprint or template used to create objects. It defines what properties (variables) and behaviors (methods) the objects will have. For example, a Car class can define attributes like color and speed.

##3. What is an object in OOP?

- An object is an instance of a class. It represents a real-world entity created using the class template. Every object has its own data, like c = Car() where c becomes an object.

##4. Difference between abstraction and encapsulation

- Abstraction focuses on hiding unnecessary details and showing only important features. Encapsulation means wrapping data and methods inside a class and restricting direct access. Abstraction is about “what”, and encapsulation is about “how”.

##5. What are dunder methods in Python?

- Dunder methods are special methods that start and end with double underscores, like __init__ and __str__. They allow us to use built-in operations on objects. For example, __add__ helps in operator overloading.

##6. Explain inheritance in OOP.

- Inheritance allows one class to take features from another class. The class that inherits is called the child, and the class providing features is the parent. It helps in code reusability. Example: Dog inheriting from Animal.

##7. What is polymorphism in OOP?

- Polymorphism means the same method name can behave differently depending on the object. This helps in writing flexible programs. Example: fly() works differently for a Sparrow and a Penguin.

##8. How is encapsulation achieved in Python?

- Encapsulation is done by making variables private using __variable. Access to data is controlled through methods like getters and setters. This protects the data from accidental modification.

##9. What is a constructor in Python?

- A constructor is a special method __init__() that runs automatically when an object is created. It initializes the object’s variables. Example: setting name and age in a Person class.

##10. What are class and static methods in Python?

- Class methods use @classmethod and can access class variables using cls. Static methods use @staticmethod and do not depend on the class or object. They behave like normal functions inside a class.

##11. What is method overloading in Python?

- Python does not support traditional method overloading. Instead, it can be simulated by giving default values or using variable arguments (*args). This allows a method to handle different numbers of inputs.

##12. What is method overriding in OOP?

- Method overriding happens when a child class provides its own version of a method already defined in the parent class. This helps achieve runtime polymorphism. Example: Dog.speak() overriding Animal.speak().

##13. What is a property decorator in Python?

- @property is used to make a method act like a variable. It helps in controlling access to class attributes without calling a method. It is commonly used for getters and setters.

##14. Why is polymorphism important in OOP?

- Polymorphism makes code more flexible and reusable because a single function can work with different object types. It improves readability and supports the idea of writing generic code.

##15. What is an abstract class in Python?

- An abstract class is a class that cannot be directly instantiated. It contains abstract methods that must be implemented by child classes. Python provides this using the abc module.

##16. What are the advantages of OOP?

- OOP makes code easier to maintain by organizing it into classes. It improves reusability through inheritance. It also enhances security with encapsulation and allows flexibility using polymorphism.

##17. What is multiple inheritance in Python?

- Multiple inheritance means a class can inherit from more than one parent class. This allows combining features from different classes. Example: class C inheriting from class A and B.

##18. Difference between class variable and instance variable

- A class variable is shared by all objects of the class. An instance variable is unique to each object. Changes in instance variables do not affect other objects.

##19. Purpose of __str__ and __repr__ methods

- __str__ returns a readable string for users, usually friendly output. __repr__ returns a detailed string meant for developers. If __str__ is missing, Python uses __repr__.

##20. Significance of super() in Python

- super() is used to call methods of the parent class. It helps in reusing parent logic without rewriting it. Commonly used in constructors.

##21. Significance of the __del__ method

- __del__ is a destructor method that runs when an object is deleted. It is used to release resources or cleanup tasks. Python automatically calls it when objects are destroyed.

##22. Difference between @staticmethod and @classmethod

- @staticmethod does not access class or instance data. @classmethod can access class-level data using cls. Class methods are used for alternative constructors.

##23. How polymorphism works in Python with inheritance?

- When a child class overrides a parent method, Python decides which method to run based on the object's type. This is runtime polymorphism. Example: different play() methods in instruments.

##24. What is method chaining in Python OOP?

- Method chaining means calling multiple methods on the same object in one line. It works when methods return the object itself. Example: obj.setA().setB().show().

##25. Purpose of the __call__ method in Python

- __call__ allows an object to be called like a function. When implemented, writing obj() executes the __call__ method. It is used in advanced programming cases.

## 1. Create a parent class Animal with a method speak() that prints a generic message. Create a child class Dog that overrides the speak() method to print "Bark!".

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Bark!")

d = Dog()
d.speak()
```

##2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14 * self.r * self.r

class Rectangle(Shape):
    def __init__(self, l, b):
        self.l = l
        self.b = b
    def area(self):
        return self.l * self.b

print(Circle(5).area())
print(Rectangle(4, 6).area())
```

##3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.

```python
class Vehicle:
    def __init__(self, type):
        self.type = type

class Car(Vehicle):
    def __init__(self, type, brand):
        super().__init__(type)
        self.brand = brand

class ElectricCar(Car):
    def __init__(self, type, brand, battery):
        super().__init__(type, brand)
        self.battery = battery

ec = ElectricCar("4-wheeler", "Tesla", "85 kWh")
print(ec.type, ec.brand, ec.battery)
```

##4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes Sparrow and Penguin that override the fly() method.

```python
class Bird:
    def fly(self):
        print("Bird is flying")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high")

class Penguin(Bird):
    def fly(self):
        print("Penguins cannot fly")

for bird in [Sparrow(), Penguin()]:
    bird.fly()
```

##5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.

```python
class BankAccount:
    def __init__(self):
        self.__balance = 0

    def deposit(self, amt):
        self.__balance += amt

    def withdraw(self, amt):
        if amt <= self.__balance:
            self.__balance -= amt
        else:
            print("Not enough balance")

    def check_balance(self):
        return self.__balance

acc = BankAccount()
acc.deposit(500)
acc.withdraw(200)
print(acc.check_balance())
```

##6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar and Piano that implement their own version of play().

```python
class Instrument:
    def play(self):
        print("Instrument is playing")

class Guitar(Instrument):
    def play(self):
        print("Guitar sound")

class Piano(Instrument):
    def play(self):
        print("Piano sound")

for i in [Guitar(), Piano()]:
    i.play()
```

##7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.

```python
class MathOperations:

    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

print(MathOperations.add_numbers(4, 6))
print(MathOperations.subtract_numbers(10, 7))
```

##8. Implement a class Person with a class method to count the total number of persons created.

##10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(1, 4)
print(v1 + v2)
```

##12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades.

```python
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        return sum(self.grades) / len(self.grades)

s = Student("Aman", [80, 90, 70])
print(s.average_grade())
```

##14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary

```python
class Employee:
    def calculate_salary(self, hours, rate):
        return hours * rate

class Manager(Employee):
    def calculate_salary(self, hours, rate, bonus):
        return super().calculate_salary(hours, rate) + bonus

m = Manager()
print(m.calculate_salary(40, 50, 500))
```

##15. Create a class Product with attributes name, price, and quantity. Implement a method total_price() that calculates the total price of the product.

```python
class Product:
    def __init__(self, name, price, qty):
        self.name = name
        self.price = price
        self.qty = qty

    def total_price(self):
        return self.price * self.qty

p = Product("Pen", 5, 10)
print(p.total_price())
```

##16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Moo")

class Sheep(Animal):
    def sound(self):
        print("Baa")

Cow().sound()
Sheep().sound()
```

##17. Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that returns a formatted string with the book's details.

##18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms

```python
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

class Mansion(House):
    def __init__(self, address, price, rooms):
        super().__init__(address, price)
        self.rooms = rooms

m = Mansion("Hill Road", 5000000, 12)
print(m.address, m.price, m.rooms)
```
